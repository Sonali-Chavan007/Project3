# Histogram

```python
A histogram is a graphical representation of the distribution of numerical data. 
It consists of a series of adjacent bars that display the frequency or count of data falling into specific intervals, known as bins.
```

```python
A Pie Chart is a circular statistical plot that can display only one series of data. 
The area of the chart is the total percentage of the given data. 
The area of slices of the pie represents the percentage of the parts of the data.
```

# Boxplot

```python
A Box Plot is also known as Whisker plot is created to display the summary of the set of 
data values having properties like minimum, first quartile, median, third quartile and maximum. 
In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. 
Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.
```

```python
sns.boxplot(x='SalePrice',y='SaleCondition',data=df, )
plt.show()
```

# Kernel Density Estimate(KDE)Plots

```python
Density plots uses Kernel Density Estimation (so they are also known as Kernel density estimation plots or KDE)
which is a probability density function. 
The region of plot with a higher peak is the region with maximum data points residing between those values.
```

```python
sns.kdeplot(df['GrLivArea'],fill=True)
plt.title("Kernel Density Estimate for GrLivArea")
plt.show()
```

# ECDF(Emprirical Cumulative Distribution function)Plots

```python
ECDF stands for Empirical Commutative Distribution. 
It is more likely to use instead of the histogram for visualizing 
the data because the ECDF plot visualizes each and every data point of the dataset directly, 
which makes it easy for the user to interact with the plot.
```

```python

sorted_values = np.sort(df['OverallQual'])
ecdf = np.arange(1, len(sorted_values) + 1) / len(sorted_values)

plt.plot(sorted_values, ecdf, marker='_', linestyle='none')
plt.title('ECDF for Overall Quality')
plt.xlabel("Overall Quality")
plt.ylabel("ECDF")
plt.show()
```

# BAR Plot

```python
A bar plot or bar chart is a graph that represents the category of data with rectangular bars
with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally
or vertically. A bar chart describes the comparisons between the discrete categories. 
One of the axis of the plot represents the specific categories being compared, while 
the other axis represents the measured values corresponding to those categories.
```

```python
sns.barplot(x='MSSubClass',y='SalePrice',data=df)
plt.title('Average sale price of each MSSubClass')
plt.xticks(rotation=90)
plt.show()
```

# Pair Plot

```python
To plot multiple pairwise bivariate distributions in a dataset, you can use the .pairplot() function. 

The diagonal plots are the univariate plots, and this displays the relationship for the (n, 2) 
combination of variables in a DataFrame as a matrix of plots.

```

```python
pair_plt = ['OverallQual', 'GrLivArea', 'TotalBsmtSF', 'GarageCars', 'SalePrice']
sns.pairplot(df[pair_plt])
plt.suptitle('Pairwise Relationship') 
plt.show()

```

# Cumulative Distribution Function For 'GrLivArea'

```python
Matplotlib is a library in Python and it is a numerical — mathematical extension for the NumPy library.  
The cumulative distribution function (CDF) of a real-valued random variable X, or 
just distribution function of X, evaluated at x, is the probability that X will take a value less than or equal to x.
```

```python
sorted_values = np.sort(df['GrLivArea'])
plt.plot(sorted_values, np.linspace(0, 1, len(sorted_values), endpoint=False))  # Corrected 'np.linespace' to 'np.linspace'
plt.title('CDF for GrLivArea')
plt.xlabel('GrLivArea')  # Corrected 'xlable' to 'xlabel'
plt.ylabel('Cumulative Probability')  # Corrected 'ylable' to 'ylabel'
plt.show()
```

# Boxen Plot

```python
Draw an enhanced box plot for larger datasets. 
This style of plot was originally named a “letter value” plot 
because it shows a large number of quantiles that are defined as “letter values”.  
It is similar to a box plot in plotting a nonparametric representation of a distribution in which 
all features correspond to actual observations. By plotting more quantiles, it provides more information about 
the shape of the distribution, particularly in the tails.

```

```python
sns.boxenplot(x="GarageArea", y="GarageQual", data=df)

plt.title('Size of Garage in car capasity by Square feet')
plt.show()
```
